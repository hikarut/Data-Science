```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

plt.show()
# fig.savefig('my_figure.png')
```

```python
# axは2つの座標軸オブジェクトの配列
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
plt.style.use('seaborn-whitegrid')

# 軸の範囲
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)

# グラフのタイトル
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")

# グラフの色
plt.plot(x, np.sin(x), color='#FFDD44', label='sin(x)') # HEX code
plt.plot(x, np.sin(x - 1), color='chartreuse', label='sin(x - 1)') # all HTML color names supported

# 凡例
plt.legend()
```

```python
plt.plot(x, x + 5, color='#FFDD44', linestyle='--')
plt.plot(x, x + 7, color='chartreuse', linestyle=':')
```

```python
# 矢印と注釈
fig, ax = plt.subplots()
x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4), arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6), arrowprops=dict(arrowstyle='->', connectionstyle="angle3,angleA=0,angleB=-90"))
```

```python
def hist_and_lines(title):
    np.random.seed(0)
    
    fig, ax = plt.subplots(1, 2, figsize=(11, 4))
    fig.suptitle(title, fontsize=12)
    
    ax[0].hist(np.random.randn(1000))
    for i in range(3):
        ax[1].plot(np.random.rand(10))
    ax[1].legend(['a', 'b', 'c'], loc='lower left')
    
with plt.style.context('default'):
    hist_and_lines('default')
with plt.style.context('ggplot'):
    hist_and_lines('ggplot')
with plt.style.context('seaborn-poster'):
    hist_and_lines('seaborn-poster')
with plt.style.context('bmh'):
    hist_and_lines('bmh')
with plt.style.context('fivethirtyeight'):
    hist_and_lines('fivethirtyeight')
```
