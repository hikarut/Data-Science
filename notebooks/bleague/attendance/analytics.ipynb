```python
import pandas as pd

b1 = pd.read_csv("b1.csv", sep=",")
b1
```

```python
b1.arena.unique()
# b1.arena.nunique()
```

```python
b1.groupby('home_team').mean()
```

```python
# win_count = b1[b1['home_victory'] == 'win'].groupby('home_team').count()['setu']
win_count = b1[b1['home_victory'] == 'win'].groupby('home_team').count()
win_count[['setu']]
```

```python
# arena_rate = b1.groupby('home_team').mean()['arena_rate']
arena_rate = b1.groupby('home_team').mean()
arena_rate[['arena_rate']]
```

```python
output = pd.merge(win_count[['setu']], arena_rate[['arena_rate']], how="inner" ,on="home_team")
# output

output2 = output.reset_index()
output2
```

```python
import japanize_matplotlib
import matplotlib.pyplot as plt

output.plot.scatter(x='setu', y='arena_rate', style='home_team')
# plt.scatter(x=output['setu'], y=output['arena_rate'])
# plt.plot(output['setu'], output['arena_rate'], marker='.')
```
