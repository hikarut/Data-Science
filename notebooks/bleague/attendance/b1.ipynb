```python
team_mapping = {
    '新潟': 1, 
    'SR渋谷': 2, 
    '京都': 3, 
    '三河': 4, 
    '琉球': 5, 
    '三遠': 6, 
    '秋田': 7, 
    '千葉': 8, 
    '北海道': 9, 
    '富山': 10, 
    '横浜': 11, 
}
arena_mapping = {
    '立川立飛': 1,
    '駒沢': 2
}
# atokyo['away_team_number'] = atokyo.apply(lambda x: f"{team_mapping[x['away_team']]}", axis=1)
# atokyo['arena_number'] = atokyo.apply(lambda x: f"{arena_mapping[x['arena']]}", axis=1)
atokyo = pd.get_dummies(atokyo)
atokyo
```

```python
from sklearn import tree

# データ
# feature = atokyo[['away_team_number', 'arena_number', 'attendance']].values
feature = atokyo[[
    'away_team_SR渋谷',
    'away_team_新潟',
    'away_team_京都',
    'away_team_三河',
    'away_team_琉球',
    'away_team_三遠',
    'away_team_秋田',
    'away_team_千葉',
    'away_team_北海道',
    'away_team_富山',
    'away_team_横浜',
#     'arena_立川立飛',
#     'arena_駒沢',
    'attendance'
]].values
feature_names = [
    'away_team_SR_shibuya',
    'away_team_新潟',
    'away_team_京都',
    'away_team_三河',
    'away_team_琉球',
    'away_team_三遠',
    'away_team_秋田',
    'away_team_chiba',
    'away_team_北海道',
    'away_team_富山',
    'away_team_横浜',
#     'arena_立川立飛',
#     'arena_駒沢',
    'attendance'
]
# target = atokyo[['home_victory']].values
target = atokyo[['home_victory_win']].values
# target_names = ['win', 'lose']
target_names = ['lose', 'win']
# print(feature)

# モデル作成
clf = tree.DecisionTreeClassifier(max_depth=3)  # limit depth of tree
clf.fit(feature, target)
```

```python
import pydotplus
from IPython.display import Image
from graphviz import Digraph

# 可視化
dot_data = tree.export_graphviz(
    clf,
    out_file=None,
    feature_names=feature_names,
    class_names=target_names,
    filled=True,
#     proportion=True
)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
from dtreeviz.trees import dtreeviz
import numpy as np

# 可視化
viz = dtreeviz(
    clf,
    feature, 
    np.array([i[0] for i in target]),
    target_name='win',
    feature_names=feature_names,
    class_names=[str(i) for i in target_names],
    orientation='TD' # TD or LR
) 

display(viz)
```
