```python
import pandas as pd

b1 = pd.read_csv("b1.csv", sep=",")
# b1.head(10)
# b1
b1[(b1['home_team'] == '横浜') & (b1['home_away'] == 'home')]
```

```python
# win_count = b1[b1['home_victory'] == 'win'].groupby('home_team').count()['setu']
win_count = b1[b1['home_victory'] == 'win'].groupby('home_team').count()
win_count[['setu']]
```

```python
# arena_rate = b1.groupby('home_team').mean()['arena_rate']
# arena_rate = b1.groupby('home_team').mean()
# arena_rate = b1[b1['home_away'] == 'home'].groupby('home_team').mean()
arena_rate = b1[(b1['home_away'] == 'home') & (b1['attendance'] != 0)].groupby('home_team').mean()
# arena_rate[['arena_rate']]
# arena_rate[['attendance']]
arena_rate[['attendance', 'arena_rate']]
```

```python
# output = pd.merge(win_count[['setu']], arena_rate[['arena_rate']], how="inner" ,on="home_team")
output = pd.merge(win_count[['setu']], arena_rate[['attendance']], how="inner" ,on="home_team")
# output

output2 = output.reset_index().rename(columns={'setu': 'victory'})
# output2.sort_values('attendance', ascending=False)
output2[(output2['home_team'] == '三遠') | (output2['home_team'] == 'A東京')]
```

```python
# 外れ値(三遠)除外
# output2_drop = output2.drop(3, axis=0)
# 外れ値(三遠,新潟)除外
# output2_drop = output2.drop(3, axis=0).drop(13, axis=0)
# 三遠、東京
# output2_drop = output2.drop(3, axis=0).drop(0, axis=0)
output2_drop = output2
output2_drop

# statsmodelsを使った回帰分析
import statsmodels.api as sm
from sklearn import linear_model

victory = output2_drop[['victory']]
# arena_rate = output2_drop[['arena_rate']]
arena_rate = output2_drop[['attendance']]

model = linear_model.LinearRegression()
model.fit(victory.values, arena_rate.values)

sm_model = sm.OLS(arena_rate, sm.add_constant(victory))
result = sm_model.fit(disp=0)
print(result.summary())
```

```python
# 回帰直線
fig, ax = plt.subplots(figsize=(16, 9))
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))

# plt.rcParams["figure.figsize"] = [15,7] # 図の縦横のサイズ([横(inch),縦(inch)])
# plt.rcParams["font.size"] = 15 # フォントサイズ デフォルト12
# plt.title("scatter")
plt.xlabel("victory")
plt.ylabel("attendance")
plt.scatter(victory, arena_rate, s=20, alpha=0.5)
plt.plot(victory.values, model.predict(victory.values), color='black')
```
