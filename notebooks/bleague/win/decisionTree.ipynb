```python
import pandas as pd

# 教師データ
b1_2017 = pd.read_csv("b1_2017.csv", sep=",")
b1_2018 = pd.read_csv("b1_2018.csv", sep=",")
b1_2017_2018 = pd.concat([b1_2017, b1_2018])
b1_all = pd.concat([b1_2017, b1_2018, b1_2019])

# 予測データ
b1_2019 = pd.read_csv("b1_2019.csv", sep=",")
```

```python
atokyo_2017_2018 = b1_2017_2018[(b1_2017_2018['home_team'] == 'A東京') & (b1_2017_2018['home_away'] == 'home')]
atokyo_2019 = b1_2019[(b1_2019['home_team'] == 'A東京') & (b1_2019['home_away'] == 'home')]
atokyo_all = b1_all[(b1_all['home_team'] == 'A東京') & (b1_all['home_away'] == 'home')]

pd.set_option('display.max_rows', 100)
# atokyo_2017_2018
# atokyo_2019
atokyo_all
```

```python
atokyo_2017_2018 = pd.get_dummies(atokyo_2017_2018)
atokyo_2019 = pd.get_dummies(atokyo_2019)
atokyo_all = pd.get_dummies(atokyo_all)
atokyo_all
```

```python
from sklearn import tree

feature = atokyo_2017_2018[[
    'away_team_SR渋谷',
    'away_team_新潟',
    'away_team_京都',
#     'away_team_三河',
    'away_team_琉球',
    'away_team_三遠',
    'away_team_秋田',
    'away_team_千葉',
    'away_team_北海道',
    'away_team_富山',
    'away_team_横浜',
#     'away_team_名古屋D',
#     'away_team_大阪',
#     'away_team_川崎',
#     'away_team_栃木',
#     'away_team_滋賀',
    'attendance',
    'home_score',
    'away_score',
]].values
feature_names = [
    'away_team_shibuya',
    'away_team_niigata',
    'away_team_kyoto',
#     'away_team_mikawa',
    'away_team_ryukyu',
    'away_team_sanen',
    'away_team_akita',
    'away_team_chiba',
    'away_team_hokkaido',
    'away_team_toyama',
    'away_team_yokohama',
#     'away_team_nagoya',
#     'away_team_osaka',
#     'away_team_kawasaki',
#     'away_team_tochigi',
#     'away_team_shiga',
    'attendance',
    'home_score',
    'away_score',
]

target = atokyo_2017_2018[['home_victory_win']].values
target_names = ['lose', 'win']

# モデル作成
clf = tree.DecisionTreeClassifier(max_depth=3)  # limit depth of tree
clf.fit(feature, target)
```

```python
import pydotplus
from IPython.display import Image
from graphviz import Digraph

# 可視化
dot_data = tree.export_graphviz(
    clf,
    out_file=None,
    feature_names=feature_names,
    class_names=target_names,
    filled=True,
#     proportion=True
)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
from dtreeviz.trees import dtreeviz
import numpy as np

# 可視化
viz = dtreeviz(
    clf,
    feature, 
    np.array([i[0] for i in target]),
    target_name='win',
    feature_names=feature_names,
    class_names=[str(i) for i in target_names],
    orientation='TD' # TD or LR
) 

display(viz)
```

```python
import numpy

# 予測
feature_2019 = atokyo_2019[[
    'away_team_SR渋谷',
    'away_team_新潟',
    'away_team_京都',
#     'away_team_三河',
    'away_team_琉球',
    'away_team_三遠',
    'away_team_秋田',
    'away_team_千葉',
    'away_team_北海道',
    'away_team_富山',
    'away_team_横浜',
#     'away_team_名古屋D',
#     'away_team_大阪',
#     'away_team_川崎',
#     'away_team_栃木',
#     'away_team_滋賀',
    'attendance',
    'home_score',
    'away_score',
]].values
target_2019 = atokyo_2019[['home_victory_win']].values

# 作成したモデルを用いて予測を実行
predicted = clf.predict(feature_2019)

# 予測結果
print(predicted)

# 予測精度
numpy.average(sum(predicted == target_2019) / len(target_2019))
```
